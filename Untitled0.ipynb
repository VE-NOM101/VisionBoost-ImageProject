```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.fft import fft2, ifft2, fftshift, ifftshift

# 1) Load grayscale image
image = cv2.imread('Lena.jpg', 0)
rows, cols = image.shape[:2]

# 2) Take FFT and shift zero frequency to center
F = fftshift(fft2(image))

# 3) Create frequency noise (spikes in spectrum)
# Choose position of spikes away from center (u,v)
u_offset, v_offset = 5, 6  # control direction and frequency
# Find maximum magnitude in shifted FFT (before adding spikes)
max_val = np.max(np.abs(F))

# Define the amplitude of the noise spikes as a percentage of max_val
percentage = 0.1  # e.g., 10%
amp = percentage * max_val

# Add symmetric noise spikes
F[rows//2 + u_offset, cols//2 - v_offset] += amp
F[rows//2 - u_offset, cols//2 + v_offset] += amp

# 4) Inverse FFT to get noisy image
noisy_image = np.abs(ifft2(ifftshift(F)))

# 5) Display
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.imshow(noisy_image, cmap='gray')
plt.title('Image with periodic noise (from freq domain)')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(np.log1p(np.abs(F)), cmap='gray')
plt.title('Modified FFT magnitude (with spikes)')
plt.axis('off')
plt.show()

cv2.imwrite('im.png', noisy_image)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.fft import fft2, ifft2, fftshift, ifftshift

# --- Step 1: Load and add periodic noise ---
image = cv2.imread('Lena.jpg', 0)
rows, cols = image.shape
F = fftshift(fft2(image))

# Add symmetric spikes
u_offset, v_offset = 50, 30
max_val = np.max(np.abs(F))
amp = 0.05 * max_val
F[rows//2 + u_offset, cols//2 - v_offset] += amp
F[rows//2 - u_offset, cols//2 + v_offset] += amp

noisy_image = np.abs(ifft2(ifftshift(F)))

# --- Step 2: Compute log magnitude spectrum ---
magnitude_spectrum = np.log1p(np.abs(F))

# --- Step 3: Mask DC component ---
mask = np.ones_like(magnitude_spectrum)
r = 10
cv2.circle(mask, (cols//2, rows//2), r, 0, -1)
spectrum_masked = magnitude_spectrum * mask

# --- Step 4: Adaptive threshold ---
mean_val = np.mean(spectrum_masked)
std_val = np.std(spectrum_masked)
T = mean_val + 4 * std_val
binary_mask = (spectrum_masked > T).astype(np.uint8)

# --- Step 5: Non-Maximum Suppression (robust version) ---
# Convert to 8-bit image for OpenCV
norm = cv2.normalize(spectrum_masked, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

# Use dilation to find local maxima
kernel = np.ones((37,37), np.uint8)
local_max = cv2.dilate(norm, kernel)

# Keep only maxima that are above threshold and equal to their local max
nms = np.zeros_like(norm, np.uint8)
nms[(norm == local_max) & (binary_mask == 1)] = 255

# --- Step 6: Extract coordinates of detected spikes ---
coords = cv2.findNonZero(nms)

# --- Step 7: Visualize ---
plt.figure(figsize=(14,5))
plt.subplot(1,3,1); plt.imshow(magnitude_spectrum, cmap='gray'); plt.title('FFT Magnitude')
plt.subplot(1,3,2); plt.imshow(binary_mask*255, cmap='gray'); plt.title('After Threshold')
plt.subplot(1,3,3); plt.imshow(nms, cmap='gray'); plt.title('After Non-Maximum Suppression')
plt.show()

if coords is not None:
    coords = [tuple(pt[0]) for pt in coords]
    print("Detected spike coordinates:", coords)
else:
    print("No spikes detected.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.fft import fft2, ifft2, fftshift, ifftshift

# --- Step 1: Load and add multiple periodic noises ---
image = cv2.imread('Lena.jpg', 0)
rows, cols = image.shape
F = fftshift(fft2(image))

max_val = np.max(np.abs(F))
amp = 0.05 * max_val  # amplitude of each noise spike

# Noise source 1 (low frequency diagonal)
u_offset1, v_offset1 = 50, 30
F[rows//2 + v_offset1, cols//2 + u_offset1] += amp
F[rows//2 - v_offset1, cols//2 - u_offset1] += amp

# Noise source 2 (different direction/frequency)
u_offset2, v_offset2 = 70, -40
F[rows//2 + v_offset2, cols//2 + u_offset2] += amp
F[rows//2 - v_offset2, cols//2 - u_offset2] += amp

# Inverse FFT  noisy image
noisy_image = np.abs(ifft2(ifftshift(F)))

# --- Step 2: Compute log magnitude spectrum ---
magnitude_spectrum = np.log1p(np.abs(F))

# --- Step 3: Mask DC region ---
mask = np.ones_like(magnitude_spectrum)
cv2.circle(mask, (cols//2, rows//2), 10, 0, -1)
spectrum_masked = magnitude_spectrum * mask

# --- Step 4: Threshold + NMS to detect spikes automatically ---
mean_val = np.mean(spectrum_masked)
std_val = np.std(spectrum_masked)
T = mean_val + 3 * std_val
binary_mask = (spectrum_masked > T).astype(np.uint8)

# Normalize for NMS
norm = cv2.normalize(spectrum_masked, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
kernel = np.ones((21,21), np.uint8)
local_max = cv2.dilate(norm, kernel)
nms = np.zeros_like(norm, np.uint8)
nms[(norm == local_max) & (binary_mask == 1)] = 255

coords = cv2.findNonZero(nms)
if coords is None:
    raise ValueError("No spikes detected!")
coords = [tuple(pt[0]) for pt in coords]

# --- Step 5: Build Butterworth Notch Reject Filter ---
def butterworth_notch_reject(shape, D0=10, n=2, notch_centers=[]):
    rows, cols = shape
    H = np.ones((rows, cols), dtype=np.float32)
    u = np.arange(rows) - rows/2
    v = np.arange(cols) - cols/2
    V, U = np.meshgrid(v, u)

    for (x, y) in notch_centers:
        u_k = y - cols/2
        v_k = x - rows/2
        Dk = np.sqrt((U - u_k)**2 + (V - v_k)**2)
        Dk_neg = np.sqrt((U + u_k)**2 + (V + v_k)**2)
        H *= 1 / (1 + ((D0**2) / (Dk * Dk_neg + 1e-6))**n)
    return H

# --- Step 6: Apply Butterworth notch filter ---
D0 = 10
n = 2
H = butterworth_notch_reject((rows, cols), D0, n, coords)
F_filtered = F * H

# --- Step 7: Inverse FFT to reconstruct image ---
recovered_image = np.abs(ifft2(ifftshift(F_filtered)))

# --- Step 8: Visualization ---
plt.figure(figsize=(15,9))
plt.subplot(2,3,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(2,3,2), plt.imshow(noisy_image, cmap='gray'), plt.title('Noisy Image (2 periodic noises)')
plt.subplot(2,3,3), plt.imshow(np.log1p(np.abs(F)), cmap='gray'), plt.title('FFT Magnitude (with multiple spikes)')
plt.subplot(2,3,4), plt.imshow(binary_mask*255, cmap='gray'), plt.title('Thresholded Peaks')
plt.subplot(2,3,5), plt.imshow(H, cmap='gray'), plt.title('Butterworth Notch Filter')
plt.subplot(2,3,6), plt.imshow(recovered_image, cmap='gray'), plt.title('Recovered Image')
plt.tight_layout()
plt.show()

cv2.imwrite('im2.png', noisy_image)

print("Detected spike coordinates:", coords)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.fft import fft2, ifft2, fftshift, ifftshift

# --- Step 1: Load and add multiple periodic noises ---
image = cv2.imread('Lena.jpg', 0)
rows, cols = image.shape
F = fftshift(fft2(image))

max_val = np.max(np.abs(F))
amp = 0.05 * max_val  # amplitude of each noise spike

# Noise source 1 (low frequency diagonal)
u_offset1, v_offset1 = 50, 30
F[rows//2 + v_offset1, cols//2 + u_offset1] += amp
F[rows//2 - v_offset1, cols//2 - u_offset1] += amp

# Noise source 2 (different direction/frequency)
u_offset2, v_offset2 = 70, -40
F[rows//2 + v_offset2, cols//2 + u_offset2] += amp
F[rows//2 - v_offset2, cols//2 - u_offset2] += amp

# Inverse FFT  noisy image
noisy_image = np.abs(ifft2(ifftshift(F)))

# --- Step 2: Compute log magnitude spectrum ---
magnitude_spectrum = np.log1p(np.abs(F))

# --- Step 3: Mask DC region ---
mask = np.ones_like(magnitude_spectrum)
cv2.circle(mask, (cols//2, rows//2), 10, 0, -1)
spectrum_masked = magnitude_spectrum * mask

# --- Step 4: Threshold + NMS to detect spikes automatically ---
mean_val = np.mean(spectrum_masked)
std_val = np.std(spectrum_masked)
T = mean_val + 3 * std_val
binary_mask = (spectrum_masked > T).astype(np.uint8)

# Normalize for NMS
norm = cv2.normalize(spectrum_masked, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
kernel = np.ones((21,21), np.uint8)
local_max = cv2.dilate(norm, kernel)
nms = np.zeros_like(norm, np.uint8)
nms[(norm == local_max) & (binary_mask == 1)] = 255

coords = cv2.findNonZero(nms)
if coords is None:
    raise ValueError("No spikes detected!")
coords = [tuple(pt[0]) for pt in coords]

# --- Step 5: Build Butterworth Notch Reject Filter ---
def butterworth(img,uk_list, vk_list, D0, n):
    M, N = img.shape
    H = np.ones((M, N), dtype=np.float32)
    for k in range(len(uk_list)):
        uk = uk_list[k]
        vk = vk_list[k]
        for u in range(M):
            for v in range(N):
                Dk = np.sqrt((u - N/2 - uk)**2 + (v - M/2 - vk)**2)
                D_k = np.sqrt((u - N/2 + uk)**2 + (v - M/2 + vk)**2)

                if Dk == 0:
                    Dk = 1e-6
                if D_k == 0:
                    D_k = 1e-6

                # Butterworth notch reject formula
                Hk = 1 / (1 + (D0 / Dk)**(2 * n))
                H_k = 1 / (1 + (D0 / D_k)**(2 * n))
                H[v, u] = H[v, u] * Hk * H_k


    return H

# --- Step 6: Apply Butterworth notch filter ---
D0 = 10
n = 2
uk_list = []
vk_list = []

for point in coords:
    t1 = 256 - point[0]
    t2 = 256 - point[1]
    uk_list.append(t1)
    vk_list.append(t2)
H = butterworth(image, uk_list,vk_list,D0,n)
F_filtered = F * H

# --- Step 7: Inverse FFT to reconstruct image ---
recovered_image = np.abs(ifft2(ifftshift(F_filtered)))

# --- Step 8: Visualization ---
plt.figure(figsize=(15,9))
plt.subplot(2,3,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(2,3,2), plt.imshow(noisy_image, cmap='gray'), plt.title('Noisy Image (2 periodic noises)')
plt.subplot(2,3,3), plt.imshow(np.log1p(np.abs(F)), cmap='gray'), plt.title('FFT Magnitude (with multiple spikes)')
plt.subplot(2,3,4), plt.imshow(binary_mask*255, cmap='gray'), plt.title('Thresholded Peaks')
plt.subplot(2,3,5), plt.imshow(H, cmap='gray'), plt.title('Butterworth Notch Filter')
plt.subplot(2,3,6), plt.imshow(recovered_image, cmap='gray'), plt.title('Recovered Image')
plt.tight_layout()
plt.show()

cv2.imwrite('im2.png', noisy_image)

print("Detected spike coordinates:", coords)
```
