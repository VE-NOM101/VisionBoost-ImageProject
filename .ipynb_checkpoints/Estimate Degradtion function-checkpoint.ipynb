```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def imshow(title, img, cmap='gray'):
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def motion_H(shape, a, b, T=1.0):
    """
    Implements H(u,v) = [T / (π(ua+vb))] * sin[π(ua+vb)] * exp(-jπ(ua+vb))
    shape: (rows, cols)
    a,b  : motion components (related to direction & length)
    """
    h, w = shape
    u = np.fft.fftfreq(w)
    v = np.fft.fftfreq(h)
    U, V = np.meshgrid(u, v)
    s = U*a + V*b
    eps = 1e-8
    H = (T / (np.pi*(s+eps))) * np.sin(np.pi*s) * np.exp(-1j*np.pi*s)
    return H
```

```python
# You can replace this path with your own Lena image
img = cv2.imread("Lena.jpg", cv2.IMREAD_GRAYSCALE)
if img is None:
    raise SystemExit("Couldn't read Lena.jpg — please place it in working dir.")

img = cv2.resize(img, (256,256))
f = img.astype(np.float32) / 255.0
imshow("Original Image", f)
```

```python
# Motion parameters
T = 1.0
L = 20           # motion length (pixels)
theta = 30       # motion angle (degrees)

# Compute components a,b
a = L * np.cos(np.deg2rad(theta))
b = L * np.sin(np.deg2rad(theta))

# Build transfer function H(u,v)
H_true = motion_H(f.shape, a, b, T)

# Apply blur: G = F * H
F = np.fft.fft2(f)
G = F * H_true
g = np.fft.ifft2(G).real
g = np.clip(g, 0, 1)

# Display results
imshow("Degraded Image (motion blur)", g)

# Display log magnitude of |H|
H_mag = np.fft.fftshift(np.log(np.abs(H_true)+1e-6))
imshow("Log Magnitude of H(u,v)", H_mag, cmap='inferno')
```
